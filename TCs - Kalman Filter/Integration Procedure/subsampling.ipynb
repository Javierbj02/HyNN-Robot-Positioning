## 3 - Subsampling to match   
Subsample the route positions to match the model prediction rate in a real-world scenario.
This sumsampling is only done to no noise scenarios, since other noise levels scenarios are built from this sumbsampled one.

```python
import pandas as pd
```

```python
def subsample_dataframe(df, keep_indices, block_size):

    mask = []
    for i in range(len(df)):
        if i % block_size in keep_indices:
            mask.append(True)
        else:
            mask.append(False)


    subsampled_df = df[mask].reset_index(drop=True)
    return subsampled_df

data_8 = pd.read_csv("Data/TC1/No noise/ULA_8.csv")
data_16 = pd.read_csv("Data/TC1/No noise/ULA_16.csv")
data_32 = pd.read_csv("Data/TC1/No noise/ULA_32.csv")
data_64 = pd.read_csv("Data/TC1/No noise/ULA_64.csv")

data_8 = subsample_dataframe(data_8, keep_indices=[0, 2, 5, 7, 10], block_size=12)  # ULA 8
data_16 = subsample_dataframe(data_16, keep_indices=[0], block_size=3)  # ULA 16
data_32 = subsample_dataframe(data_32, keep_indices=[0], block_size=3)  # ULA 32
data_64 = subsample_dataframe(data_64, keep_indices=[0], block_size=4)  # ULA 64

data_8.to_csv("Data/TC1/No noise/ULA_8.csv", index=False)
data_16.to_csv("Data/TC1/No noise/ULA_16.csv", index=False)
data_32.to_csv("Data/TC1/No noise/ULA_32.csv", index=False)
data_64.to_csv("Data/TC1/No noise/ULA_64.csv", index=False)
```
